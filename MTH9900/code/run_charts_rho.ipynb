```python
import math
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

import pickle
```

```python
with open ('euput_rBergomi_results_rho', 'rb') as fp:
    euput_rBergomi_results_rho = pickle.load(fp)
with open ('amput_rBergomi_results_rho', 'rb') as fp:
    amput_rBergomi_results_rho = pickle.load(fp)
with open ('euput_localvol_results_rho', 'rb') as fp:
    euput_localvol_results_rho = pickle.load(fp)
with open ('amput_localvol_results_rho', 'rb') as fp:
    amput_localvol_results_rho = pickle.load(fp)
with open ('Klist_rBergomi_rho', 'rb') as fp:
    Klist_rBergomi_rho = pickle.load(fp)
with open ('Klist_localvol_rho', 'rb') as fp:
    Klist_localvol_rho = pickle.load(fp)
```

```python
euput_rBergomi_results_rho = np.array(euput_rBergomi_results_rho)
amput_rBergomi_results_rho = np.array(amput_rBergomi_results_rho)
euput_localvol_results_rho = np.array(euput_localvol_results_rho)
amput_localvol_results_rho = np.array(amput_localvol_results_rho)
```

```python
eep_rBergomi_results_rho = amput_rBergomi_results_rho - euput_rBergomi_results_rho
eep_localvol_results_rho = amput_localvol_results_rho - euput_localvol_results_rho
eep_differential_rho = eep_rBergomi_results_rho - eep_localvol_results_rho
```
